```python
A = torch.rand(8, 3, 4)
B = torch.rand(8, 4, 5)
print(A @ B)
print(A[0,] @ B[0, ])
```

```python
print(torch.equal(A[0,] @ B[0, ], (A @ B)[0]))
print(torch.allclose(A[0,] @ B[0, ], (A @ B)[0]))
```

## view() never copies data, share memory

```python
x = torch.tensor([[1, 2, 3],
                         [4, 5, 6]])
y = x.view(3, 2)
y[0,0] = 99
print(x[0])
```

```python
x = torch.arange(12).view(3,4).t()
print(x.is_contiguous())
print(x[0])
y = x.view(4, 3)
y[0,0] = 99
print(x[0])
```

```python
x = torch.rand(3,4)
y = x.T
print(x)
y[0]=0
print(x)
print(y)
```

## Creating a new tensor from a transpose using `.clone()`

When you use `x.T`, it returns a *view* of the original tensor `x` with its dimensions permuted. This means that both `x` and `x.T` share the same underlying memory. If you want to create a completely *independent* transposed tensor, you need to explicitly create a copy using the `.clone()` method.

```python
x = torch.rand(3, 4)
print("Original tensor x:")
print(x)

y = x.T.clone() # Creates a new tensor y, which is a clone of the transpose of x
print("\nNew tensor y (cloned transpose of x):")
print(y)

# Modify y and observe that x is unaffected
y[0, 0] = 99.0
print("\nTensor y after modification (y[0,0] = 99.0):")
print(y)

print("\nTensor x after y was modified (x should be unchanged):")
print(x)
```
