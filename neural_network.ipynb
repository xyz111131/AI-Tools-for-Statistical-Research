```python
import torch
```

# Construct a multilayer percepton

```python
class NeuralNetwork(torch.nn.Module):
    def __init__(self, num_inputs, num_outputs):
        super().__init__()

        self.layers = torch.nn.Sequential(

            # 1st hidden layer
            torch.nn.Linear(num_inputs, 30),
            torch.nn.ReLU(),

            # 2nd hidden layer
            torch.nn.Linear(30, 20),
            torch.nn.ReLU(),

            # output layer
            torch.nn.Linear(20, num_outputs),
        )

    def forward(self, x):
        logits = self.layers(x)
        return logits
```

# Instantiate a new neural network

```python
model = NeuralNetwork(50, 3)
print(model)
```

## check the number of trainable parameters

```python
num_params = sum(
    p.numel() for p in model.parameters() if p.requires_grad
)
print("Total number of trainable model parameters:", num_params)
```

## access weight and bias of each layer

```python
print(model.layers[0].weight) # requires_grad = True by default
print(model.layers[0].bias)
```

```python
torch.manual_seed(123) # set seed for reproducibility
```

# Model inference/Forward pass

```python
X = torch.rand((1, 50))
out = model(X) # forward pass
print(out) # grad_fn: the last-used function to compute a variable in the computational graph, Addmm: matrix multiplication (mm) followed by an addition
```

```python
with torch.no_grad():  # no need to keep track of gradients for model inference
    out = model(X)
print(out)
```

```python
with torch.no_grad():  # output class probabilities
    out = torch.softmax(model(X), dim=1)
print(out)
```
